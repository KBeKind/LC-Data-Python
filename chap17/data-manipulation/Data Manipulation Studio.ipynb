## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!
import pandas as pd
import numpy as np

baltimore = pd.read_csv('dataset/baltimore_9-24-2016_9-30-2017.csv')
boston = pd.read_csv('dataset/boston_9-24-2016_9-30-2017.csv')
new_york = pd.read_csv('dataset/new-york_9-24-2016_9-30-2017.csv')
philadelphia = pd.read_csv('dataset/philadelphia_9-24-2016_9-30-2017.csv')

dataframes = {"baltimore":baltimore, "boston":boston,"new_york": new_york,"philadelphia": philadelphia}

for (key, value) in dataframes.items():
    display(value.head())
```

## Clean Your Data

In the last lesson, we cleaned the data to related to San Francisco. Pull up your notebook from the last lesson and use your cleaning skills to clean the dataframes as necessary.

```python
# Clean your data here!

for(key, value) in dataframes.items():
    for col in value.columns:
        percentage_missing = np.mean(value[col].isnull())
        print(f"{key} column: {col.title()} has {percentage_missing*100}% missing")
    print("***********************************************")


print("***********************************************")

for (key, value) in dataframes.items():
    total_cells = np.product(value.shape)
    missing_cells = pd.isnull(value).sum()
    total_missing = missing_cells.sum()
    percentage_missing = round((total_missing/total_cells), 2) *100
    print(f"{key} percent missing: {percentage_missing}%")   


clean_dfs = {}
    
for(key, df) in dataframes.items():
    columns_to_keep = []
    for col in df.columns:
        percentage_missing = np.mean(df[col].isnull())
        if(percentage_missing < 0.5):
            columns_to_keep.append(col)
    cleaned_df = df[columns_to_keep].copy()       
    clean_dfs[f"{key}"] = cleaned_df

print("***********************************************")

    
for (key, value) in clean_dfs.items():
    total_cells = np.product(value.shape)
    missing_cells = pd.isnull(value).sum()
    total_missing = missing_cells.sum()
    percentage_missing = round((total_missing/total_cells), 2) *100
    print(f"{key} percent missing: {percentage_missing}%")   


    
for(key, df) in clean_dfs.items():
    for col in df.columns:
        if df[col].dtype in ['int64', 'float64']:
            mean_value = df[col].mean()
            df[col].fillna(mean_value, inplace=True)

for(key, df) in clean_dfs.items():
    for col in df.columns:
        mode_value = df[col].mode().iloc[0] if not df[col].mode().empty else None
        if mode_value is not None:
            df[col].fillna(mode_value, inplace=True)
 
    
print("***********************************************")

    
for (key, value) in clean_dfs.items():
    total_cells = np.product(value.shape)
    missing_cells = pd.isnull(value).sum()
    total_missing = missing_cells.sum()
    percentage_missing = round((total_missing/total_cells), 2) *100
    print(f"{key} percent missing: {percentage_missing}%")       
    
for(key, value) in clean_dfs.items():
    for col in value.columns:
        percentage_missing = np.mean(value[col].isnull())
        print(f"{key} column: {col.title()} has {percentage_missing*100}% missing")
    print("***********************************************")
    

print(clean_dfs["baltimore"]["Item Size"])
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
combined_df = pd.concat(clean_dfs.values(), ignore_index=True)

print(combined_df)
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of unit of sale in the Northeast region? In the last lesson, we learned that a unit of sale could be something like a bin or individually. 
2. What is the average number of pumpkins for each variety that came into terminal markets for the year by region? Pumpkin varieties include Howden and Fairytale pumpkins.

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.

# combined_df["Variety"]
grouped_df = combined_df.groupby('Variety')
mean_low_price = grouped_df['Low Price'].mean()
low_low_price = grouped_df['Low Price'].min()
max_low_price = grouped_df['Low Price'].max()

print(f"LOWEST LOW PRICE:\n {low_low_price}")
print("*********************************************")

print(f"MAX LOW PRICE:\n{max_low_price}")
print("*********************************************")

print(f"MEAN LOW PRICE BY VARIETY:\n{mean_low_price}")
print("*********************************************")
print("*********************************************")

mean_high_price = grouped_df["High Price"].mean()
low_high_price = grouped_df['High Price'].min()
max_high_price = grouped_df['High Price'].max()

print(f"LOWEST HIGh PRICE:\n {low_high_price}")
print("*********************************************")

print(f"MAX HIGH PRICE:\n{max_high_price}")
print("*********************************************")
print(f"MEAN HIGH PRICE BY VARIETY:\n{mean_high_price}")
print("*********************************************")
```

```python
# Put your code here to find the average number of pumpkins coming into terminal markets of each variety.
total_sales_per_variety = combined_df['Variety'].value_counts()
print(total_sales_per_variety)


combined_df['Date'] = pd.to_datetime(combined_df['Date'], format='%m/%d/%Y')
combined_df['Year'] = combined_df['Date'].dt.year
counts_per_variety_per_city_per_year = combined_df.groupby(['City Name', 'Variety', 'Year']).size()
sales_counts_df = counts_per_variety_per_city_per_year.reset_index(name='Yearly Sales:')

# Print the result
print(sales_counts_df)

```
